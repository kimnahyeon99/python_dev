```python
# 2. 기본적인 get 요청
# get(url) : 해당 URL에 HTTP GET 요청을 보낸다
# status_code : 응답 코드(200[정상처리],400[요청잘못],500[서버에러]) 
# text : 웹페이지를 문자열로 반환
import requests

# 크롤링한 웹페이지 URL 저장
url = 'https://www.naver.com' # 메인 페이지

# get 요청
response = requests.get(url)

# 응답 코드 확인
print('응답 상태 코드 : ', response.status_code)

# 페이지 내용 출력(html 코드)
print('='*100)
print(response.text)
print('='*100)
print('단어수 : ',len(response.text))
```

```python
# 요청 헤어(User-Agent) 변경하기
# 사유 : 일부 웹사이트는 User-Agent가 없는 요청을 차단할 수 있다
# 일부 사이트는 기본적인 봇 요청을 차단한다.
# 서버입장에서 코드로 실행하는 것이 아니고 브라우저에서 접속한 것처럼 보이도록 설정

import requests

url = 'https://naver.com'

# 요청 헤더 설정(일반적인 브라우저 사용하는 것처럼 설정)
headers = { 
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36" 
}

# GET 요청
response = requests.get(url,headers=headers)

# response 상태 확인, 텍스트 추출
print('응답 상태 코드 : ', response.status_code)
print(response.text[:100])
```

```python
# 구글 검색 
# https://www.google.com/search?q=%EB%94%A5%EB%9F%AC%EB%8B%9D&rlz=1C5MACD_enKR1134KR1134&oq=%EB%94%A5%EB%9F%AC%EB%8B%9D&gs_lcrp=EgZjaHJvbWUyDggAEEUYJxg5GIAEGIoFMgcIARAAGIAEMgcIAhAAGIAEMgcIAxAAGIAEMgcIBBAAGIAEMgcIBRAAGIAEMgcIBhAuGIAEMgcIBxAAGIAEMgcICBAAGIAEMgcICRAAGIAE0gEKMjU1MzJqMGoxNagCALACAA&sourceid=chrome&ie=UTF-8
url = 'https://www.google.com/search'

# get 요청에 사용할 파라미터 설정 : 딕셔너리 형태 설정
params = {'q':'딥러닝'}
headers = { 
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36" 
}

# 요청 보내기 : get
response = requests.get(url, params=params) # url 뒤에 파라미터 붙는다는 뜻

print(response.text)
```

```python
# naver 검색
query = '머신러닝'
url = f'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query={query}'


# get 요청에 사용할 파라미터 설정 : 딕셔너리 형태 설정
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
}


# 요청 보내기 : get
response = requests.get(url,  headers=headers)


print(response.text)
```

```python
# 5. post 방식 요청 처리하기 : insert, update, delete
# requests.post(url,data) : data를 post 방식으로 서버에 넘긴다
# url : 요청 주소
# data : 서버로 보낼 데이터
# response.json() : 서버에서 json 방식으로 넘어온 데이터를 딕셔너리로 변환 반환

url = 'https://httpbin.org/post'

# 보낼 데이터
data = {
    'username' : 'testuser',
    'password' : 'testpassword'
}

# post 요청
response = requests.post(url, data)
print('응답 코드 : ', response.status_code)
print(response.json()) # REST API 방식에서 서버에서 넘어온 값 JSON
```

```python
# 6. JSON 데이터 요청 및 처리
# 증권사 api, REST Api => 응답이 데이터 온다 (json)

# 게시판 1번글 요청 => json 응답
url = 'https://jsonplaceholder.typicode.com/posts/1'

response = requests.get(url)

# 여기까지 reponse는 결과가 json 형태이긴 하지만 문자열

# 위 문자열을 딕셔너리로 변환
data = response.json()

print('제목 : ', data['title'])
print('내용 : ', data['body'])
```

```python
# 크롤링시 예외처리 및 오류 대응 방법
# 이 코드는 실행 안됨
url = 'http://example.com'

# 예외 처리
try:
    response = requests.get(url, timeout=5) # 서버가 응답을 안하면 5초까지만 기달리겠다는 뜻 // 5초 동안 응답을 받겠다
    response.raise_for_status() # 4xx, 5xx 응답시 예외를 발생시키겠다. 
    print(response.text[:100])
except requests.Timeout: # 타입아웃 되었을 때 처리
    print('요청 시간이 초과')
except requests.HTTPError as e: # 네트워크 오류
    print('HTTP 오류 : ',e)
except requests.RequestException as e: # 진짜 요청 중 오류
    print('요청 중 오류 발생 : ',e)
```
